```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
x = ["a","b","c","d"]
y = [20,30,70,90]
plt.plot(x,y, color=("red"))
plt.show()
```

```python
xdata = np.array([0,6])
ydata = ([0,250])
plt.plot(xdata,ydata,marker=("*"))
plt.show()
```

```python
#ms-marker size
a = [3,8,1,10]
b = ["a","b","c","d","e","f"]
plt.plot(a,marker ="o",ms="20")
```

```python
plt.figure(figsize=(4,4))
xpoints = np.array([1,8])
ypoints = np.array([3,10])
plt.plot(xpoints, ypoints, "*")
plt.show()
```

```python
x = [10,20,80,40,50]
y = [30,70,10,60,70]
z = ["a","b","c","d","e"]
plt.plot(z,x,label=("2025-sales"))
plt.plot(z,y, label=("2026-sales"))
plt.legend()
plt.show()
```

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])
font1 ={'family':'sarif','color':'blue','size':20}
font2 ={'family':'sarif','color':'darkred','size':15}
plt.title("sport watch data",fontdict=font1)
plt.xlabel("Avarage pulse",fontdict=font2)
plt.ylabel("calorie burnage",fontdict=font2)
plt.grid()
plt.grid(axis="y")
plt.grid(color="green",linestyle="--",linewidth="0.5")
plt.plot(x,y)
plt.show()
```
